```python
import pandas as pd
```

```python
# Realizar esse processo para cada dataset da pasta uploads
arquivo_csv = ('dataclean/analise_txcg_jan_2025.csv')
df = pd.read_csv(arquivo_csv, sep='#', encoding='utf-8-sig')
print(df.shape)
```

```python
df.head()
```

```python

filtro = (
    (df['comarca'] == 'APARECIDA DE GOIÂNIA') & 
    (df['serventia'] == '1ª Vara Cível') &
    (df['data_distribuicao'])
)
df[filtro].head()
```

```python
print(df[filtro].shape)
```

```python
baixados = df[filtro]['data_baixa'].notna().sum()
print(baixados)
```

```python
pendentes = df.loc[filtro, 'data_baixa'].isna().sum()
print(pendentes)
```

```python
tx_cong_global = (100 * pendentes/(pendentes+baixados)).round(2)
print(f"Taxa de Congestionamento Global: {tx_cong_global} %")
```

```python
dist = df[filtro]['data_distribuicao'].notna().sum()
print(dist)
```

```python
df['data_distribuicao'] = pd.to_datetime(df['data_distribuicao'], errors='coerce') 

dist_2025 = (df[filtro]['data_distribuicao'].dt.year == 2025).sum()
print(dist_2025)
```
